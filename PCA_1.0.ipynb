```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df=pd.read_excel('testovaci_dataset_BP.xlsx', sheet_name='Sheet1') 
df.sample(5) #vypisuje n random radku 
```

```python
df.columns
```

```python
col=[ 'Soudni spory', 'Status', 'Pocet let na trhu',
       'Rust zak. kapitalu', 'Pocet jinych subjektu',
       'Jine subjekty v lividace', 'Status domeny', 'Pravni forma',
       'Pocet zamestnancu', 'Rust poctu zam.',
       'Nespolehlivy platce', 'Pocet jednatelu', 'Ochranne znamky']
```

```python
pd.options.mode.chained_assignment = None 
df[col].fillna(0, inplace=True) 
```

```python
df[col].corr()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaler.fit(df[col])
```

```python
scaled_data=scaler.transform(df[col])
```

```python
scaled_data
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca=pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(15,10))
plt.scatter(x_pca[:,0],x_pca[:,1],c=df['Pocet let na trhu'])
plt.xlabel('First principle component')
plt.ylabel('Second principle component')
```
